```python
from numpy import pi, cos

def faktorial (n):
  f = 1
  for i in range(1,n+1):
    f = f*i
  return f

sudut = int(input('Masukan Sudut :'))
n = int(input('Masukan Jumlah Perulangan :'))

x=sudut*pi/180
#s = x
c = 1
sign = -1
sa = cos(x)
print("nilai cos{:d} metode analitik adalah {:.20f}.".format(sudut,sa))
for i in range (2,n+1):
  if i%2 == 0:
    c = c + sign*((x**i)/faktorial(i))
    sign = -1*sign
  Err = abs (c-sa)
  print("nilai cos{:d} adalah {:.20f} dengan error {:.20f}.".format(sudut,c,Err))
```
